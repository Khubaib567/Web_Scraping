```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install()) 
driver.get('https://mor.nlm.nih.gov/RxNav/search')

ndc_value=['00006494300']

scroll_down=driver.find_element_by_id('searchById')
select_value = Select(scroll_down)


select_value.select_by_visible_text('NDC')
enter_value=driver.find_element_by_id('searchTerm')
enter_value.send_keys(ndc_value)
enter_value.send_keys(Keys.ENTER)
i_tag= driver.find_elements_by_class_name("fa-clipboard")





```
